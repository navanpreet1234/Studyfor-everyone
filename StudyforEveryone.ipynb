```python
import pyttsx3#pip install pyttsx3
import speech_recognition as sr#pip install speechRecognition
from playsound import playsound#pip install playsound
import pandas as pd#pip install pandas
from word2number import w2n#pip install word2number
from csv import writer
l=[]#creating empty list
#intializing engines voice
engine = pyttsx3.init('sapi5')
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id)
#reading data form csv file
data=pd.read_csv('Data.csv')
data.dtypes
#Defining speak function for engine.
def speak(audio):
    engine.say(audio)
    engine.runAndWait()
#Defining takeCommand function for taking command from user via microphone.
def takeCommand():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        r.adjust_for_ambient_noise(source, duration=1)
        r.pause_threshold = 1
        audio = r.listen(source)
    try:
        print("Listening")
        query = r.recognize_google(audio, language='en-in')
        print(f"User said: {query}\n")

    except Exception as e:
        speak("Say that again please...")
        query = takeCommand()
    return query
#Defining course function forregistring new users and provinf lessons to already registered users
def course(z):
    if z==0:
        speak("Please give the foll details")
        speak("Tell you Name")
        n=takeCommand()
        l.append(n)
        speak("Tell you Age")
        l.append(w2n.word_to_num(takeCommand()))
        speak("Tell you Gender")
        l.append(takeCommand())
        speak("Tell you Class")
        a1=w2n.word_to_num(takeCommand())
        l.append(a1)
        speak("Tell your City")
        l.append(takeCommand())
        speak("Tell your Monthly income")
        l.append(w2n.word_to_num(takeCommand()))
        z=data['Id'].tail(1).item()+1
        l.append(z)
        with open('Data.csv','a+',newline='') as file:
            csv_writer=writer(file)
            csv_writer.writerow(l)
        speak("Please note that your user id is")
        speak(z)
        speak("Thank you for registration please call again and continue your course using your id")
   
    elif z==1:  
        speak("Tell your name")
        n=takeCommand()
        speak("Tell your User id")
        b=w2n.word_to_num(takeCommand())
        if data[(data['Name']==n) & (data['Id']==b)].shape[0] == 1:
            a=data['Class'].loc[(data['Name']==n) & (data['Id']==b)]
            a1=a.item()
            speak("To study english say English and to study math say Maths")
            q=takeCommand().lower()
            speak("If you want to study from chapter 1 say 1 else say the chapter number")
            y=w2n.word_to_num(takeCommand())
            l1=['ch1ej.mp3','ch3ej.mp3','ch3ej.mp3','ch4ej.mp3']
            l2=['ch1mj.mp3','ch2mj.mp3','ch3mj.mp3','ch4mj.mp3']
            l3=['ch1es.mp3','ch2es.mp3','ch3es.mp3','ch4es.mp3']
            l4=['ch1ms.mp3','ch2ms.mp3','ch3ms.mp3','ch4ms.mp3']
            if q=='english':
                if a1 in [1,2,3]:
                    for i in range(y,2):
                        playsound(l1[i-1])
                elif a1 in ['4','5','6']:
                    for i in range(y,2):
                        playsound(l2[i-1])
            elif q=='maths':
                if a1 in [1,2,3]:
                    for i in range(y,2):
                        playsound(l3[i-1])

                elif a1 in [4,5,6]:
                    for i in range(y,2):
                        playsound(l4[i-1])
    else:
        speak("You are not a registered user. Please register yourself")
if __name__ == "__main__":
    speak("If this is your first time say 0 else say 1")
    z=w2n.word_to_num(takeCommand())
    course(z)
   
    data.head()  
```
